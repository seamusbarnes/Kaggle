```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join('Learning Agency Lab - Automated Essay Scoring 2.0/data', 'leaderboard_05_04.csv'))

def plot_leaderboard():
    fig = plt.figure(figsize=(9, 3))
    plt.plot(df['Score'])
    plt.xlim((0, len(df)))
    plt.ylim((0, 1))
    plt.xlabel('Leaderboard position')
    plt.ylabel('Score')
    plt.title('Leaderboard')
    plt.show()
plot_leaderboard()

def plot_score_distribution():
    fig = plt.figure(figsize=(9,3))
    plt.hist(df['Score'].clip(0.7, 1), bins=[i for i in np.linspace(0.70, 0.82, 121)])
    plt.xlabel('Score')
    plt.ylabel('Occurences')
    plt.title('Distribution of scores (lower bound clipped at 0.7')
    plt.show()
plot_score_distribution()
```
